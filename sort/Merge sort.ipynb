```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def merge_sort(input_array, key=-1):
    size = len(input_array)
    if size > 1:
        mid = size // 2
        left_array = input_array[:mid].copy()
        right_array = input_array[mid:].copy()
        
        merge_sort(left_array, key)
        merge_sort(right_array, key)
        
        p = 0
        q = 0
        r = 0
        
        left_size = len(left_array)
        right_size = len(right_array)
        
        while p < left_size and q < right_size:
            if key == -1:
                if left_array[p] < right_array[q]:
                    input_array[r] = left_array[p]
                    p += 1
                else:
                    input_array[r] = right_array[q]
                    q += 1
            else:
                if left_array[p][key] < right_array[q][key]:
                    input_array[r] = left_array[p]
                    p += 1
                else:
                    input_array[r] = right_array[q]
                    q += 1
                
            r += 1
            
        while p < left_size:
            input_array[r] = left_array[p]
            p += 1
            r += 1
            
        while q < right_size:
            input_array[r] = right_array[q]
            q += 1
            r += 1
    
```

```python
%%time
test_array = random.sample(range(50000), 5000)
print(test_array)
```
